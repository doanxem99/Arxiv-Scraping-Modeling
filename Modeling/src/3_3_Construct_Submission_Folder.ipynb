## Construct Submission Folder Structure

```python
import os
import shutil
from pathlib import Path

print("Define libraries")
```

```python
# Configuration
BASE_DIR = Path("/mnt/d/Programming/School/NMKHDL/Lab2")
PAPERS_DIR = BASE_DIR / "papers"
BIBTEX_DIR = BASE_DIR / "bibtex"
STUDENT_ID = "23127088"

# Get all publication folders from bibtex/STUDENT_ID
bibtex_student_dir = BIBTEX_DIR / STUDENT_ID
publication_folders = [f for f in bibtex_student_dir.iterdir() if f.is_dir()]

# Copy files for each publication
for pub_folder in publication_folders:
    pub_id = pub_folder.name
    
    # Rename <pub_id>.json to hierarchy.json
    old_bibtex_json = pub_folder / f"{pub_id}.json"
    if old_bibtex_json.exists():
        old_bibtex_json.rename(pub_folder / "hierarchy.json")
    
    # Copy files from papers/pub_id/ to bibtex/STUDENT_ID/pub_id/
    paper_folder = PAPERS_DIR / pub_id
    
    if paper_folder.exists():
        metadata_json = paper_folder / "metadata.json"
        references_json = paper_folder / "references.json"
        
        if metadata_json.exists():
            shutil.copy2(metadata_json, pub_folder / "metadata.json")
        
        if references_json.exists():
            shutil.copy2(references_json, pub_folder / "references.json")

    # Delete cleaned_data.json if it exists
    cleaned_data_json = pub_folder / "cleaned_data.json"
    if cleaned_data_json.exists():
        cleaned_data_json.unlink()

    

print(f"Files copied to: {bibtex_student_dir}")
print(f"Total publications: {len(publication_folders)}")
```
